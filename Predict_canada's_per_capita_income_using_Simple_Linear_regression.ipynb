```python
import pandas as pd
import numpy as np
#from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/a.csv')
df
```

```python
new_df = df.drop('Income',axis='columns')
#new_df = df.area
new_df
```

```python
Income = df.Income
Income 
```

```python
# Create linear regression object
reg = LinearRegression()
reg.fit(new_df, Income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y= new_df *reg.coef_ + reg.intercept_
y
```

```python
plt.xlabel('Year')
plt.ylabel('Income')

plt.scatter(new_df ,Income,color='blue',marker='+')
plt.plot(new_df,y,color='red',marker='+')
```
